# Q1: What is multithreading in Python? Why is it used? Name the module used to handle threads in Python. Why is the  threading module used? Write the use of the following functions (activeCount, currentThread, enumerate).

Multithreading is a concurrent execution of multiple threads (smaller units of a process) within the same process. It allows the processor to execute multiple threads simultaneously, enhancing the performance of applications that require parallelism.

In Python, the threading module is used to handle threads. The threading module provides a way to create and manage threads in a high-level manner.

activeCount(): Returns the number of Thread objects currently alive.

currentThread(): Returns the current Thread object.

enumerate(): Returns a list of all Thread objects currently alive.

# Q3: Explain the following functions (run, start, join, isAlive).

run(): This method is called when the start() method of a Thread is invoked. It should be overridden in a subclass to implement the specific behavior of the thread.

start(): Initiates the thread's activity. It must be called at most once per thread object. It initializes the thread and invokes the run() method.

join([timeout]): Waits for the thread to finish. The optional timeout argument specifies the maximum time the join() method should wait for the thread to terminate.

isAlive(): Returns True if the thread is alive (has been started and not terminated), and False otherwise.

import threading

def print_squares():
    for i in range(1, 6):
        print(f"Square of {i}: {i*i}")

def print_cubes():
    for i in range(1, 6):
        print(f"Cube of {i}: {i*i*i}")


thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)


thread1.start()
thread2.start()


thread1.join()
thread2.join()
# Q5: State advantages and disadvantages of multithreading.

Advantages:

Improved performance by utilizing multiple processors or cores.
Responsiveness in GUI applications, as the main thread remains responsive to user interactions while background tasks are executed in separate threads.
Efficient handling of I/O-bound operations, as one thread can continue execution while others are waiting for I/O.
Disadvantages:

Complexity: Multithreading introduces complexity, making the code harder to write, understand, and debug.
Potential for race conditions and deadlocks: Synchronization issues can occur when multiple threads access shared resources simultaneously.
Global Interpreter Lock (GIL): In CPython, the Global Interpreter Lock limits the execution of multiple threads in parallel, affecting the performance of CPU-bound tasks.
Q6: Explain deadlocks and race conditions.

Deadlock: A deadlock occurs when two or more threads are blocked forever, each waiting for the other to release a resource. In other words, each thread holds a resource and is waiting for another resource acquired by another thread. Deadlocks can lead to the system coming to a standstill.

Race Condition: A race condition occurs when the behavior of a program depends on the relative timing of events, such as the order of execution of threads. It arises when multiple threads access shared resources concurrently, and the final outcome depends on the order of execution. Race conditions can lead to unpredictable behavior and bugs in a program. Synchronization mechanisms, such as locks, are used to avoid race conditions.
